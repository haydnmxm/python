```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
from numpy import random
```

### 1.Complete a program to plot the scatter of dependence points of two factors as shown in the figure

```python
x = range(100)
y = range(100) + np.random.randint(0,30,100)
plt.scatter(x, y)
plt.title('Simple Scatter plot')
plt.xlabel('X - value')
plt.ylabel('Y - value')
plt.show()
```

### 2.Plot the sine and cosine functions as shown in the figure

```python
x1 = np.arange(0,3*np.pi,0.1) 
s = np.sin(x1)
c = np.cos(x1)

plt.plot(x1,s,x1,c)
plt.show()
```

### 3. Seaborn is a Python data visualization library based on Matplotlib that provides a high-level interface for drawing attractive and informative statistical graphics, so plot the normal distribution using the seaborn library as shown in the figure


```python
fig = sns.histplot(x, color="skyblue", label="x", kde=True)
fig = sns.histplot(y, color="red", label="y", kde=True)

plt.legend() 
plt.show()
```

### 4. Lets use the distribution using a kernel density estimation, which Seaborn does with sns.kdeplot. You can use the same data set, called Data, as in the previous example 


```python
fig = sns.kdeplot(x, shade=True, color="r")
fig = sns.kdeplot(y, shade=True, color="b")
plt.show()
```

```python
data = {'Name': ['Ahmed', 'Omar', 'Ali', 'Ziad', 'Salwa', 'Lila'], 'Mobile_Sales': [2540, 1370, 1320, 2000, 2100, 2150], 'TV_Sales': [2200, 1900, 2150, 1850, 1770, 2000]}  
df = pd.DataFrame(data)
df['General_Sales'] = df['Mobile_Sales'] + df['TV_Sales']
df
```

### 5.Write a Python script to create a data frame for the following table:


```python
#A Create a bar plot of the sales volume
ax = df.plot.bar(x='Name', y='General_Sales', rot=0)
```

```python
#B Create a pie chart of item sales.
sales = ['tv_sales', 'mobile_sales']
tv = df['TV_Sales'].sum() 
mobile = df['Mobile_Sales'].sum() 
series = pd.Series([tv, mobile], 
                   index=sales, 
                   name='Pie chart of item sales.')
series.plot.pie(figsize=(6, 6))
```

```python
#C Create a box plot of item sales.
boxplot = df.boxplot(column=['Mobile_Sales', 'TV_Sales']) 
```

```python
#D Create an area plot of item sales.
Area = {"mobile_sales": [2540, 1370, 1320, 2000, 2100, 2150], 'tv_sales': [2200, 1900, 2150, 1850, 1770, 2000]}
Sellers = ('Ahmed', 'Omar', 'Ali', 'Ziad', 'Salwa', 'Lila' )
dataFrame   = pd.DataFrame(Area, index=Sellers);
dataFrame.plot(kind='area', stacked=False)
plt.show(block=True);
```

```python
#E Create a stacked bar plot of item sales.
Df = pd.DataFrame({'mobile_sales': [2540, 1370, 1320, 2000, 2100, 2150],
                   'tv_sales': [2200, 1900, 2150, 1850, 1770, 2000]}, index=Sellers)
ax = Df.plot.bar(stacked = True)
```
